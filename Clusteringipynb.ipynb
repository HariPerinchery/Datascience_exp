```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Ingest data

```python
wine_df = pd.read_csv('wine-clustering.csv')
```

Data Analysis

```python
wine_df.shape
```

```python
wine_df.head(5)
```

```python
wine_df.describe()
```

Modelling

```python
wcss = []

for i in range(1,6):
  km = KMeans(n_clusters=i)
  km.fit_predict(wine_df)
  wcss.append(km.inertia_)
```

```python
plt.plot(range(1,6), wcss)
```

Clustering with number of clusters = 4

```python
#convert df into a numpy array
X = wine_df.iloc[:,:].values
```

```python
km = KMeans(n_clusters=4)
y_means = km.fit_predict(wine_df)
y_means
```

```python
X[y_means == 0,0]
```
